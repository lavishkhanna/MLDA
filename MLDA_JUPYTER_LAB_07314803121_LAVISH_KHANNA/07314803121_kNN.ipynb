### ***1. Use the knn implementation from scikit learn to classify the Iris dataset. Also show how performance of the model varies with different values of k.***

***A. ACCURACY vs NUMBER OF NEIGHBORS***



```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Step 2: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Implement KNN using scikit-learn's KNeighborsClassifier
def knn_classification(X_train, y_train, X_test, y_test, k):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy

# Step 4 and 5: Evaluate the performance of the model with different values of k
k_values = range(1, 51)  # trying k values from 1 to 50
accuracies = []

for k in k_values:
    acc = knn_classification(X_train, y_train, X_test, y_test, k)
    accuracies.append(acc)

# Step 6: Plot a graph showing the accuracy variation with different k values
plt.plot(k_values, accuracies, marker='o')
plt.title('Accuracy vs. Number of Neighbors (K)')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Step 2: Split the dataset into training, validation, and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Step 3: Implement KNN using scikit-learn's KNeighborsClassifier
def knn_classification(X_train, y_train, X_val, y_val, k):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    val_error = 1 - knn.score(X_val, y_val)  # Calculate validation error
    return val_error

# Step 4 and 5: Evaluate the performance of the model with different values of k
k_values = range(1, 51)  # trying k values from 1 to 50
validation_error_rates = []

for k in k_values:
    val_error_rate = knn_classification(X_train, y_train, X_val, y_val, k)
    validation_error_rates.append(val_error_rate)

# Step 6: Plot a graph showing the validation error rate variation with different k values
plt.plot(k_values, validation_error_rates, marker='o')
plt.title('Validation Error Rate vs. Number of Neighbors (K)')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Validation Error Rate')
plt.grid(True)
plt.show()
```

###***2. Write your own implementation for knn with 4 different distance metrics and check on Iris dataset.***

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

class CustomKNN:
    def __init__(self, k=3, distance_metric='euclidean'):
        self.k = k
        self.distance_metric = distance_metric

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X_test):
        predictions = [self._predict(x) for x in X_test]
        return np.array(predictions)

    def _predict(self, x):
        # Compute distances between x and all examples in the training set
        if self.distance_metric == 'euclidean':
            distances = [np.linalg.norm(x - x_train) for x_train in self.X_train]
        elif self.distance_metric == 'manhattan':
            distances = [np.sum(np.abs(x - x_train)) for x_train in self.X_train]
        elif self.distance_metric == 'chebyshev':
            distances = [np.max(np.abs(x - x_train)) for x_train in self.X_train]
        elif self.distance_metric == 'minkowski':
            p = 3 # value of p for Minkowski distance
            distances = [np.sum(np.abs(x - x_train)**p)**(1/p) for x_train in self.X_train]

        # Sort by distance and return indices of the first k neighbors
        k_indices = np.argsort(distances)[:self.k]

        # Extract the labels of the k nearest neighbor training samples
        k_nearest_labels = [self.y_train[i] for i in k_indices]

        # Return the most common class label among the k nearest neighbors
        most_common = np.argmax(np.bincount(k_nearest_labels))
        return most_common

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Step 2: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the distance metrics to be used
distance_metrics = ['euclidean', 'manhattan', 'chebyshev', 'minkowski']

# Dictionary to store accuracy for each distance metric
accuracy_dict = {}

# Loop through each distance metric and compute accuracy
for distance_metric in distance_metrics:
    custom_knn = CustomKNN(k=3, distance_metric=distance_metric)
    custom_knn.fit(X_train, y_train)
    y_pred_custom = custom_knn.predict(X_test)
    accuracy_custom = accuracy_score(y_test, y_pred_custom)
    accuracy_dict[distance_metric] = accuracy_custom

# Print accuracy for each distance metric
for metric, accuracy in accuracy_dict.items():
    print(f'Accuracy using {metric} distance metric: {accuracy:.2f}')
```
